```python
from chromadb import HttpClient
from chromadb.utils.embedding_functions import GoogleGenerativeAiEmbeddingFunction
import pandas as pd
```

```python
chroma_client = HttpClient(host="localhost", port=5002)
collection = chroma_client.get_or_create_collection(name="file-manager")
```

```python
documents = [
        "The theory of relativity fundamentally changed our understanding of space and time, showing that they are interwoven into a single continuum known as spacetime.",
        "Machine learning models, particularly deep neural networks, have revolutionized artificial intelligence by enabling computers to learn complex patterns from large datasets.",
        "The Renaissance was a period of great cultural and intellectual change in Europe, marking the transition from the Middle Ages to modernity.",
        "Quantum computing leverages quantum-mechanical phenomena like superposition and entanglement to perform calculations that are infeasible for classical computers.",
        "The human brain contains approximately 86 billion neurons, each forming thousands of connections with other neurons, creating an incredibly complex network.",
    ]
    
metadatas = [
    {"book_name": "Physics of the Universe", "chapter": 5, "type": "textbook", "date_created": "2025-10-27T10:30:00Z"},  # Monday October 27, 2025
    {"book_name": "AI Revolution", "chapter": 3, "type": "research_paper", "date_created": "2025-10-15T14:45:00Z"},
    {"book_name": "European History", "chapter": 7, "type": "textbook", "date_created": "2025-10-05T09:15:00Z"},
    {"book_name": "Quantum Computing Fundamentals", "chapter": 4, "type": "research_paper", "date_created": "2025-10-20T16:20:00Z"},
    {"book_name": "Neuroscience Today", "chapter": 2, "type": "journal_article", "date_created": "2025-10-10T11:10:00Z"},
]

collection.upsert(
    documents=documents,
    metadatas=metadatas,
    ids=[f"doc_{i+1}" for i in range(len(documents))]
)
```

```python
results = collection.query(
    query_texts=["computers"],
    n_results=3,
    include=["metadatas", "documents", "distances"]
)

# Create a list of dictionaries for each result
result_data = []
for i, (doc, meta, dist) in enumerate(zip(
    results['documents'][0],
    results['metadatas'][0],
    results['distances'][0]
), 1):
    result_data.append({
        'Result #': i,
        'Distance': f"{dist:.4f}",
        'Document': doc,
        'Source': meta['book_name'],
        'Type': meta['type'].replace('_', ' ').title(),
        'Chapter': meta['chapter'],
        'Date Created': meta['date_created']
    })

# Create and display the DataFrame
df_results = pd.DataFrame(result_data)
df_results
```
